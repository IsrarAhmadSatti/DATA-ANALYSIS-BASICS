```python
arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])
```

```python
print(arr[0,1]) # first row second col
print(arr[1,2]) # second row tird col
print(arr[2,0]) # third row first col
```

```python
print(arr[-1,-2]) # last row sec last col
print(arr[-2,-3]) # sec last row first col
print(arr[-3,-1]) # first row last col
```

```python
# General form : arr[start:stop:step]
print(arr[0:2, 1:3]) #rws o-1, col 1-2
```

```python
print(arr[0:2, 0:2]) #rws o-1, col 0-1
```

```python
#jump of two rows and two cols
#will print first element in first row and jump to 3rd row first col
# then will jump to third row first element and then to last lement last row
print(arr[::2, ::2])
```

```python
slice_arr = arr[0, :2]
slice_arr[0] = 999
print(arr[0])
```

```python
copy_arr = arr[1].copy()
copy_arr[1] = 555
print(arr[1])
```

```python
rng = np.random.default_rng()
flips = rng.integers(0, 2, size = 100000)
prob_heads = np.mean(flips)
print("Estimated Proability of heads is:",prob_heads)
```

```python
rng = np.random.default_rng(seed=40)
flips = rng.integers(0, 2, size = 100)
prob_heads = np.mean(flips)
print(flips)
print("Estimated Proability of heads is:",prob_heads)
```
