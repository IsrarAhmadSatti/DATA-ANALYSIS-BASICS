```python
df =pd.read_csv('Cleanable.csv',sep = '\t')
```

```python
df['Country'] = df['Country'].astype('category')
```

```python
#Sorting Data by age in asceding order

df_sort = df.sort_values(by = ['Age','Country'], ascending = [True,False])
```

```python
#Joining ang Merging Data

customers = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'Name' : ['John', 'Jane', 'Joe']
})

orders = pd.DataFrame({
    'OrderID' : [101, 102, 103, 104],
    'CustomerID' : [1,2,2,3],
    'Amount' : [250, 150, 100, 300]
})
```

```python
# Inner Merge on CustomerID

df_inner = pd.merge(customers, orders, on = 'CustomerID', how = 'inner')
print("Inner Merge\n", df_inner)
```

```python
# Left Merge
df_left = pd.merge(customers, orders, on = 'CustomerID', how = 'left')
print("Left Merge\n", df_left)
```

```python
# Concatinate the Data
df_conc= pd.concat([customers, customers])
print("Concatinated datae\n", df_conc)
```

```python
df = pd.read_csv('ecommerce_sales_analytics_5000.csv', parse_dates = ['order_date'])
```

```python
df.info()
```

```python
#map function
#applied on a single column
#fast

df['paytype'] = df['payment_method'].map({
    'Card': 'Digital',
    'Wallet': 'Digital',
    'COD': 'Cash'
})
```

```python
df['paytype'].unique()
```

```python
df[['paytype','payment_method']].head()
```

```python
df.groupby('paytype').size()
```

```python
#Apply function
#can be applied on multiple columns
#slower then map

def rev_cat(row):
    if row['revenue'] > 500:
        return 'High'
    if row['revenue'] >= 200:
        return 'Medium'
    else:
        return 'Low'

df['rev_cat'] = df.apply(rev_cat, axis = 1)
```

```python
df['rev_cat'].value_counts()
```

```python
df[['rev_cat','revenue']].sample(10)
```

```python
def add_profit(data):
    data['profit'] = data['revenue'] * 0.20
    return data

def addNetrev(data):
    data['net_rev'] = data['revenue'] - data['profit']
    return data

df = (
    df.pipe(add_profit).pipe(addNetrev)
)
```
