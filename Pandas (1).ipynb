```python
data = {
    'Name': ['Alice', 'Bob', 'joe','Ross'],
    'Age': [25, 30, 22, 24],
    'city': ['paris', 'oslo', 'london', 'NY'],
    'salary': [5000, 6000, 8000, 10000]
}
df = pd.DataFrame(data)

print("Original data Frame:\n\n", df)
```

```python
# label based Selection

subset_loc = df.loc[1:3, ['Name', 'salary']]
print("Subset with .loc:\n\n", subset_loc)
```

```python
subset_iloc = df.iloc[0:2, 1:3]
print("Subset with .iloc:\n\n", subset_iloc)
```

```python
# Boolen Filtering
high_sal = df[df['salary'] > 7500]
print("High Salary Filter\n",high_sal)
```

```python
#Using .loc
df.loc[df['Age'] < 25, 'salary'] = 8000
print("low age salary change:\n", df.loc[df['Age'] < 25, "salary"])
```

```python
df = pd.read_csv(
    "retail.csv",
    parse_dates=["Brand"],
    dtype={
        'Country': 'category',
        'City': 'category'
    }
)
df.info()
#print(df.columns)
```

```python
clean_df = df[df["Price"] > 100]
subset = clean_df[['Name', 'Brand', 'Price', 'Category']]
```

```python
subset.to_excel('subset_sales.xlsx', sheet_name = 'Sales')

subset.to_json('subset_sales.json',
               orient = 'records',
               date_format = 'iso')
print('Files Exported')
subset.info()
```

```python
df = pd.read_csv(
    "Details.csv",sep='\t',
    parse_dates = ["JoinDate","DateOfBirth"],
    dtype={
        "Active": 'category',
        'Age':'category'
    }
)
df.info()
```

```python
#can convert a clumn to numeric if it is read as a string mistakenly

df['ID'] = pd.to_numeric(df['ID'])
df.info()
```

```python
#can also declare any column as we want to do after the data is read

df['Acive'] = df['Active'].astype('category')
df.info()
print(df['Active'].memory_usage(deep=True))
```

```python
df = pd.read_csv("Cleanable.csv",sep='\t',)
```

```python
df.info()
```

```python
df['Country'] = df['Country'].astype('category')
df.info()
```

```python
df.head()
```

```python
#Null Values in the data set
#printing null valeus in every column

print(df.isna().sum())
```

```python
#replacing null vales with nknown for strng data type


df['Email'] = df['Email'].fillna("Unknown")
df['Country'] = df['Country'].fillna(df['Country'].mode()[0])
```

```python
#replacing null vales with mode for integer type data

df['Age'] = pd.to_numeric(df['Age'])
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
```
