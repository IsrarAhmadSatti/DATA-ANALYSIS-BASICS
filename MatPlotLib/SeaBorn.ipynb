```python
# Data Correlation and heatmaps

df = pd.read_csv("healthcare_patient_analytics_seaborn.csv")

num_df = df[['length_of_stay_days', 'treatment_cost', 'recovery_score', 'readmission_risk']]

corr_matrix = num_df.corr()
corr_matrix
```

```python
# Heatmap - Basic Visualization
# map will show that these vales selected data set has no proper correlation

sns.heatmap(
    corr_matrix,
    annot = True,
    cmap = "coolwarm"
)
plt.title("Correlation HeatMap (healthcare DATA)")
plt.show()
```

```python
import numpy as np
```

```python
# Masking uoper triangle
# we will only show one side of the heatmap as everyvalue is repeating there
# for this we use mask

mask = np.triu(np.ones_like(corr_matrix, dtype = bool))
#This is often used when plotting a correlation heatmap (e.g., with seaborn) to avoid showing duplicate values

sns.heatmap(
    corr_matrix,
    mask= mask, #passing the mask
    annot = True,
    cmap = "coolwarm"
)
plt.title("Correlation HeatMap (Masked)")
plt.show()
```

```python
# Scaling and Normalization
df = pd.read_csv("telecom_customer_churn_feature_engineering.csv")
```

```python
#inspecting row values
df[['monthly_income', 'monthly_bill']].head()
```

```python
# wayward and skewed data
df[['monthly_income', 'monthly_bill']].describe()
```

```python
# Zscore scaling standardization
# 1 Calculate mean & Std

income_mean = df['monthly_income'].mean()
income_std = df['monthly_income'].std()

print("Mean: ",income_mean)
print("Std: ",income_std)
```

```python
# 2 apply Z Score Formula
# (value - mean ) / standard deviation

df['income_zscore'] = (
    df['monthly_income'] - income_mean) / income_std

df[['monthly_income','income_zscore']].head()
```

```python
# verifying zscore scaling
df['income_zscore'].mean(), df['income_zscore'].std()
```

```python
# Min Max Scalling Normalization (manual)

bill_min = df['monthly_bill'].min()
bill_max = df['monthly_bill'].max()

print("Min:", bill_min)
print("Max:", bill_max)
```

```python
# applying min max formula
# (value - min) / (max - min)

df['bill_minmax'] = (
    df['monthly_bill'] - bill_min) / (bill_max - bill_min)

df[['monthly_bill','bill_minmax']].head()
```

```python
df['bill_minmax'].min(), df['bill_minmax'].max()
```

```python
# show minium, maximum, and a mid value together
df.loc[
    [df['monthly_bill'].idxmin(),
     df['monthly_bill'].idxmax(),
     df['monthly_bill'].sort_values().index[len(df)//2]],
    ['monthly_bill','bill_minmax']
    ]
    
```

```python
# side by Side Comparison

df[['monthly_income','income_zscore', 'monthly_bill', 'bill_minmax']].head()
```

```python
# data Leakage -- Wrong (scaling before split)

df_scaledWrong = (df['monthly_income'] - df['monthly_income'].mean())/ df['monthly_income'].std()
```

```python
#Transforming Data

df = pd.read_csv("telecom_customer_churn_feature_engineering.csv")

df[['monthly_income', 'monthly_bill']].describe()
```

```python
# The data is right skewed in both columns
# varify

sns.set(style = 'whitegrid')
plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df['monthly_income'], kde = True, bins = 30)
plt.title("Monthly Income Distribution")

plt.subplot(1, 2, 2)
sns.histplot(df['monthly_bill'], kde = True, bins = 30)
plt.title("Monthly Bill Distribution")

plt.tight_layout()
plt.show()
```

```python
# to resolve right skewness we use log transformatonn

df['income_log'] = np.log1p(df['monthly_income'])
df[['monthly_income','income_log']].head()
```

```python
#plotting both these values side by side to see if anything changes
#you will see the transformed data is in standard position

sns.set(style = 'whitegrid')
plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df['monthly_income'], kde = True, bins = 30)
plt.title("Monthly Income Distribution")

plt.subplot(1, 2, 2)
sns.histplot(df['income_log'], kde = True, bins = 30)
plt.title("Log Transformed Data")

plt.tight_layout()
plt.show()
```

```python
# Reolving Outliers value problem
# clipping

#removing top 1% data from datbase
upper_limit = df['monthly_bill'].quantile(0.99)

df['bill_clipped'] = df['monthly_bill'].clip(upper = upper_limit)

df.loc[
    [df['monthly_bill'].idxmax(),
     df['monthly_bill'].sort_values().index[len(df)//2]],
    ['monthly_bill', 'bill_clipped']
    ]
```

```python
plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df['monthly_bill'], kde = True, bins = 30)
plt.title("Monthly Bill (Original)")

plt.subplot(1, 2, 2)
sns.histplot(df['bill_clipped'], kde = True, bins = 30)
plt.title("Monthly Bill (clipped at 99th quantile)")

plt.tight_layout()
plt.show()
```

```python
# winsorization function

from scipy.stats.mstats import winsorize
# winsorize clipp only top 10% data

df['bill_winsor'] = winsorize(
    df['monthly_bill'],
    limits = [0.01, 0.01] #[lower_cap, upper_cap]
)

df.loc[
    [df['monthly_bill'].idxmin(),
     df['monthly_bill'].idxmax()],
    ['bill_winsor']
    ]
```

```python
plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 2)
sns.histplot(df['bill_winsor'], kde = True, bins = 30)
plt.title("Monthly Bill (winsorize)")

plt.tight_layout()
plt.show()
```
