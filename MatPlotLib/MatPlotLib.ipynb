```python
import pandas as pd
```

```python
df = pd.read_csv(
    "ecommerce_sales_analytics_5000.csv",
    parse_dates = ['order_date'])
df = df.sample(n = 100)
```

```python
monthly_rev = (
    df.set_index("order_date")['revenue']
    .resample("ME").sum()
)
fig, ax = plt.subplots(figsize = (7, 4))
ax.plot(monthly_rev.index, monthly_rev.values, linewidth = 2)

ax.set_title("Monthly Revenue Trend")
ax.set_xlabel("Month")
ax.set_ylabel("Total revenue")
```

```python
fig, ax = plt.subplots(figsize = (6, 4))
ax.scatter(df['unit_price'], df['revenue'])

ax.set_title("Unit Price vs Revenue")
ax.set_xlabel("Unit Price")
ax.set_ylabel("revenue")
```

```python
# Bar Charts 
df = pd.read_csv(
    "ecommerce_sales_analytics_5000.csv",
    parse_dates = ['order_date'])
```

```python
cat_revenue = df.groupby('product_category')['revenue'].sum()
fig, ax = plt.subplots(figsize = (6, 4))
ax.bar(cat_revenue.index, cat_revenue.values)

ax.set_title("Total Revenue by Product Category")
ax.set_xlabel("Category")
ax.set_ylabel("Total revenue")
```

```python
# Getting revenue by product category and region combined
grouped = df.groupby(['product_category','region'])['revenue'].sum().unstack()

grouped.plot(kind = "bar", figsize = (10,4))
plt.title("Revenue by Category and Region")
plt.ylabel("Revenue")
plt.show()
```

```python
# Getting revenue by product category and region combined
# Stacked Data Plots

grouped.plot(kind = 'bar', stacked= True, figsize = (8, 4))
plt.title("Stacked Revenue by Category and Region")
plt.ylabel("Revenue")
plt.show()
```

```python
# HistoGrams
import numpy as np
groupA = np.random.normal(50, 10, 1000)
groupB = np.random.normal(60, 12, 1000)

plt.hist(groupA, bins = 30, density = True,
         alpha = 0.6, label = "Group A")
plt.hist(groupB, bins = 30, density = True,
         alpha = 0.6, label = "Group B")

plt.xlabel("Score")
plt.ylabel("Density")
plt.title("Distribution Comparison")
plt.legend()
plt.show()
```

```python
# Tweaking Data in our Excel Sheet

df = df.sample( n= 1000)
fig, ax = plt.subplots(figsize = (12, 4))
ax.hist(df['revenue'], bins = 30)

ax.set_title(" Reenue Distribution")
ax.set_xlabel("Revenue")
ax.set_ylabel("frequency")
```

```python
#Changing Density of the plot
#countinh how many values fall under each bin

fig, ax = plt.subplots(figsize = (12, 4))
ax.hist(df['revenue'], bins = 30, density = True, alpha = 0.6)

ax.set_title(" Reenue Distribution")
ax.set_xlabel("Revenue")
ax.set_ylabel("Density")
```

```python
# Plotting 2 histograms together
#comparing original revenue and revenue after discount

discount_rev = df['revenue'] * (1-df['discount'])
fig, ax = plt.subplots(figsize = (10, 4))
ax.hist(df['revenue'], bins = 30, alpha = 0.6, label = 'Original revenue')
ax.hist(discount_rev, bins = 30, alpha = 0.6, label = 'Discounted Revenue')

ax.legend()
ax.set_title("Revenue Comparion")
```

```python
# Making Subplots ina Plot

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, axs = plt.subplots(2, 2, figsize = (8, 6))

#line plot
axs[0, 0].plot(x, y, color = 'b')
axs[0, 0].set_title('Line Plot')

#scatter plot
axs[0, 1].scatter(x, y, color = 'r')
axs[0, 1].set_title('Scatter Plot')

axs[1, 0].hist(np.random.randn(200), bins = 20, color = 'g')
axs[1, 0].set_title('Histogram')

category = ['A', 'B', 'C', 'D']
values = [3, 5, 7, 9]
axs[1, 1].bar(category, values, color = 'm')
axs[1, 1].set_title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
# Subplots in our Data
# applying the above concepts in real data

fig, axes = plt.subplots(2, 2, figsize = (12, 6))

axes[0, 0].hist(df['revenue'], bins = 30, color = 'g')
axes[0, 0].set_title('Revenue Distribution')

#monthly revenue trend
monthly = (df.set_index('order_date')['revenue']
    .resample('ME').sum()
          )
axes[0, 1].plot(monthly.index, monthly.values, color ='r')
axes[0, 1].set_title('Monthly Revenue Trend')

#Delivery Days
axes[1, 0].hist(df['delivery_days'], bins = 10, color = 'm')
axes[1, 0].set_title('Delivery days Distribution')

#Customer rating
axes[1, 1].hist(df['customer_rating'], bins = 10, color = 'y')
axes[1, 1].set_title('Customer Rating Distribution')
```

```python
# Customization in our plots

x = np.linspace(0, 10, 20)
y = np.sin(x)

fig, ax = plt.subplots(figsize = (6, 4))
ax.plot(x, y, label = 'sin(x)', color = 'r', linewidth = 2)

#customizing ticks (resolution) and rotation
ax.set_xticks(range(0, 11, 2))
plt.xticks(rotation = 30)

#legend placement
ax.legend(loc = 'upper right')

#label and Titles
ax.set_title("customized Sin Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# saving the figure
plt.tight_layout()
plt.savefig("Sine_plot.png", dpi = 300, format = 'png')
plt.show()
```

```python
# applying in our dataset

fig, ax = plt.subplots()
ax.bar(
    df['payment_method'].value_counts().index,
    df['payment_method'].value_counts().values
)

ax.set_xlabel("Payment Method")
ax.set_ylabel("Number of Orders")
ax.tick_params(axis = 'x', rotation = 30)

plt.show()
```

```python
#Error bars
#Mean revenue by Product Category

stats = df.groupby('product_category')['revenue'].agg(['mean', 'std'])
fig, ax = plt.subplots()

ax.bar(
    stats.index,
    stats['mean'],
    yerr = stats['std'],
    capsize = 5
)

ax.set_title("Average Revenue by Category (with Variability)")
ax.set_ylabel("revenue")


```

```python
df = df.head( n = 500)
monthly = (
    df.set_index("order_date")['revenue']
    .resample("ME")
    .agg(['mean', 'std'])
)

fig, ax = plt.subplots(figsize = (10, 6))
ax.plot(monthly.index, monthly['mean'], label = 'mean revenue')

ax.fill_between(
    monthly.index,
    monthly['mean'] - monthly['std'],
    monthly['mean'] + monthly['std'],
    alpha = 0.3
)

ax.tick_params(axis = 'x', rotation = 45)
ax.set_title("Monthly Revenue with Confidence Band")
ax.legend()
    
```
